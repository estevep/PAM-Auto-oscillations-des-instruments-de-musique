```python
import numpy as np
import scipy.io.wavfile as wavfile
import librosa
import matplotlib.pyplot as plt
```

```python
# Load the audio file
audio_file = r"C:\Users\audas\OneDrive\Bureau\ATIAM\PAM\Code Repo\PAM-Auto-oscillations-des-instruments-de-musique\Descripteurs\samples\clarinet_synth_VDP_1.wav"
y, sr = librosa.load(audio_file)

# Calculate the spectrogram
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)

# Calculate the spectral centroid
spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)[0]

# Create a time array
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)

f0 = librosa.yin(y, fmin = 50,fmax = 4000)
#spectral_centroids_normalized = normalize(spectral_centroids)

# Plot the spectrogram
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')

# Plot the spectral centroid over time
plt.subplot(2, 1, 2)
plt.plot(t, f0/spectral_centroids, color='b')
plt.title('Spectral Centroid Evolution')
plt.ylabel('Normalized Spectral Centroid')
plt.xlabel('Time (s)')
plt.grid(True)

plt.tight_layout()
plt.show()



```
